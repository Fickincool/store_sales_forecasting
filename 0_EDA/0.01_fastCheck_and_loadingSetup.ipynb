Now let's read each file, print some useful info and list its columns and datatypes:

```python
def summarize_csv_file(file):
    df = pd.read_csv(file)
    print('Name: ', file.split('/')[-1])
    print('Shape of the dataframe: ', df.shape)
    print('Column and dtypes:\n', df.dtypes)
    display(df.head())
    print('\nDescription of numeric columns:\n')
    display(df.select_dtypes(exclude='object').describe())
    print('\nDescription of non-numeric columns:\n')
    try:
        display(df.select_dtypes(include='object').describe())
    except ValueError:
        print('No non-numeric columns found')
    print('\n\n')
    return
```

```python
for file in csv_files:
    summarize_csv_file(file)
```

Now lets just make a quick check of the dates in our data:

```python
def checkDates_csv_file(file):
    df = pd.read_csv(file)
    print('Name: ', file.split('/')[-1])
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df.date, errors='coerce', format='%Y-%m-%d')
        display(df.select_dtypes(include='datetime').describe(datetime_is_numeric=True))
        
    else:
        print('No "date" column.')
    print('\n\n')
    return
```

```python
for file in csv_files:
    checkDates_csv_file(file)
```
